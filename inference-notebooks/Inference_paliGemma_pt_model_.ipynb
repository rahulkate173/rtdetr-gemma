```python
from transformers import PaliGemmaForConditionalGeneration , AutoProcessor
import torch
from PIL import Image
from transformers import pipeline
```

```python
## taking model
model_id = "google/paligemma-3b-mix-224"
hf_token = "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"
model = PaliGemmaForConditionalGeneration.from_pretrained(model_id)
processor = AutoProcessor.from_pretrained(model_id)
print("[INFO] Model and processor loaded !")
```

```python
processor?
```

```python
import requests
url = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/transformers/tasks/car.jpg?download=true"
image = Image.open(requests.get(url, stream=True).raw)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
prompt = "caption this image"
model_inputs = processor(text=prompt,images=image,return_tensors="pt").to(model.device)
```

```python
model_inputs
```

```python
input_len = model_inputs["input_ids"].shape[-1]
input_len
```

```python
with torch.inference_mode():
  generation = model.generate(**model_inputs,do_sample=False,max_new_tokens=100)
  generation = generation[0][input_len:] # that much contains the image and prompt only
  decode = processor.decode(generation,skip_special_tokens=True)
  print(decode)
```
